```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def gray_gaussian_PrewittFilter(image): #-> grayscale -> gaussian filter -> prewitt filter
  image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #grayscale
  image = cv2.GaussianBlur(image, (3, 3), 0) #gaussian blur
  convolved = np.zeros(image.shape)
  G_x = np.zeros(image.shape)
  G_y = np.zeros(image.shape)
  size = image.shape
  x = np.array(([-1, 0, 1], [-1, 0, 1], [-1, 0, 1]))
  y = np.array(([-1, -1, -1], [0, 0, 0], [1, 1, 1]))
  for i in range(1, size[0] - 1): #Prewitt filter calculation
    for j in range(1, size[1] - 1):
      G_x[i, j] = np.sum(np.multiply(image[i - 1 : i + 2, j - 1 : j + 2], x))
      G_y[i, j] = np.sum(np.multiply(image[i - 1 : i + 2, j - 1 : j + 2], y))
      convolved = np.sqrt(np.square(G_x) + np.square(G_y))
      convolved = np.multiply(convolved, 255.0 / convolved.max())
      convolved = convolved.astype('uint8')
      
  return convolved
```

```python
image = cv2.imread("frut.jpg") #read image
dim=(200,200)
rescaled_image= cv2.resize(image,dim,interpolation = cv2.INTER_AREA)
Prewittfilter = gray_gaussian_PrewittFilter(rescaled_image)
f, axarr = plt.subplots(1,2,figsize=(12,12))
#axarr[0].imshow(image)
axarr[0].imshow(rescaled_image,cmap = "gray")
axarr[1].imshow(Prewittfilter,cmap = "gray")

plt.show()
```
